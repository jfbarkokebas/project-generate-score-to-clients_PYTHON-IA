```python
#Passo 1: Importar a base de dados
import pandas as pd

tabela = pd.read_csv("clientes.csv")
display(tabela)
```

```python
#Passo 2: preparar a base de dados para a inteligencia artificial
    # nesse caso especifico a base ja foi preparada, exceto os tipos

#tabela.info()

    # Modelos de IA não entendem textos , apenas numeros
    # logo precisamos transformar as colunas em textos para numeros, são elas:

        # profissão
        # mix_credito
        # comportamento_pagamento

        # Label encoder: atribuir numeros às palavras. Ex.:
            # PROFISSÃO:
                # mecanico = 1
                # encanador = 2
                # medico = 3
                # etc ...

    #importando scikit-learn:
from  sklearn.preprocessing import LabelEncoder

codificador1 = LabelEncoder()
tabela["profissao"] = codificador1.fit_transform(tabela["profissao"])

codificador2 = LabelEncoder()
tabela["mix_credito"] = codificador1.fit_transform(tabela["mix_credito"])

codificador3 = LabelEncoder()
tabela["comportamento_pagamento"] = codificador1.fit_transform(tabela["comportamento_pagamento"])

tabela.info()

# dividindo a tabela para que a IA possa aprender

    # primeira divisão:
        # y= quem eu quero prever
        # x = quem vou usar para prever


x = tabela.drop(["score_credito", "id_cliente"], axis=1)
y = tabela["score_credito"]


    #segunda divisao:

from sklearn.model_selection import train_test_split

# train_test_split --> divide a base de dados em treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y,  test_size=0.3, random_state=1)

```

```python
#Passo 3: criar um modelo de IA -> Nota/Score de cŕeditos: Bom. Ok, Ruim
    #3 passos: importa, cria, treina

    # importando modelos:
from sklearn.ensemble import RandomForestClassifier #--> arvore de decisao
from sklearn.neighbors import KNeighborsClassifier # --> vizinhos próximos

    # criar a IA:
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

    # treinar a IA:
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```
